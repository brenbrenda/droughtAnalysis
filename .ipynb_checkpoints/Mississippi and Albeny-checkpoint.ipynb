```python
#Import the monthly rainfall data(from January 1901 to December 1970) for two locations:
# 1. mydata
#Albeny_data  = pd.read_csv("rainfull.csv") 


# Load data (replace with your actual data source)
df = pd.read_csv('rainfull_Albeny.csv')
monthly_precip = df.groupby(['YEAR', 'MONTH'])['Rainfull'].sum().reset_index()

monthly_precip['Datetime'] = pd.to_datetime(monthly_precip[['YEAR', 'MONTH']].assign(DAY=1))
monthly_precip['Datetime'] = monthly_precip['Datetime'].dt.strftime('%d-%m-%Y')


formatted_df = monthly_precip[['YEAR', 'MONTH', 'Datetime', 'Rainfull']]
formatted_df.to_csv('Albeny.csv', index=False)



# Group by year and month, then sum the precipitation
albeny_data = pd.read_csv('Albeny.csv')

# Load data Missipi
df = pd.read_csv('rainfull_Mississippi.csv')
monthly_precip = df.groupby(['YEAR', 'MONTH'])['Rainfull'].sum().reset_index()

monthly_precip['Datetime'] = pd.to_datetime(monthly_precip[['YEAR', 'MONTH']].assign(DAY=1))
monthly_precip['Datetime'] = monthly_precip['Datetime'].dt.strftime('%d-%m-%Y')


formatted_df = monthly_precip[['YEAR', 'MONTH', 'Datetime', 'Rainfull']]
formatted_df.to_csv('Mississippi.csv', index=False)


# 2. Mississippi
mississippi_data = pd.read_csv("Mississippi.csv")
```

```python
#implement with SVR prediction
np.random.seed(0)
X = albeny_data.index.values.reshape(-1,1) 
y = albeny_data['Rainfull'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

svr_model = SVR(kernel='rbf')
svr_model.fit(X_train, y_train)

y_pred = svr_model.predict(X_test)

print("R2 Score:", r2_score(y_test, y_pred))

plt.figure(figsize=(16,5))
plt.plot(np.arange(len(X_test))+1, y_pred, 'go-', label="Predicted")
plt.plot(np.arange(len(X_test))+1, y_test, 'co-', label="Actual")
plt.legend()
plt.xlabel('Time')
plt.ylabel('Precipitation')
plt.title('Albeny Rainfall (SVR)')
plt.show()
```

```python
#SVR with mississipi
X = mississippi_data.index.values.reshape(-1,1)
y = mississippi_data['Rainfull'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

svr_model = SVR(kernel='rbf')
svr_model.fit(X_train, y_train)

y_pred = svr_model.predict(X_test)

print("R2 Score:", r2_score(y_test, y_pred))

plt.figure(figsize=(16,5))
plt.plot(np.arange(len(X_test))+1, y_pred, 'go-', label="Predicted")
plt.plot(np.arange(len(X_test))+1, y_test, 'co-', label="Actual") 
plt.legend()
plt.title('Mississippi Rainfall (SVR)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Assuming 'albeny_data' is your DataFrame and it's correctly set up.
np.random.seed(0)
X = albeny_data.index.values.reshape(-1, 1)  # Feature array
y = albeny_data['Rainfall'].values  # Target array

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

# Creating a simple ANN model
model = Sequential([
    Dense(64, input_dim=1, activation='relu'),  # Input layer with 64 neurons, ReLU activation
    Dense(64, activation='relu'),               # Hidden layer with 64 neurons, ReLU activation
    Dense(1)                                    # Output layer with 1 neuron (output layer for regression)
])

# Compiling the model
model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error')

# Training the model
model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=0)

# Predicting with the model
y_pred = model.predict(X_test).flatten()

# Evaluating the model
print("R2 Score:", r2_score(y_test, y_pred))

# Plotting predictions and actual values
plt.figure(figsize=(16, 5))
plt.plot(np.arange(len(X_test))+1, y_pred, 'go-', label="Predicted")
plt.plot(np.arange(len(X_test))+1, y_test, 'co-', label="Actual")
plt.legend()
plt.xlabel('Time')
plt.ylabel('Precipitation')
plt.title('Albeny Rainfall (ANN)')
plt.show()
```

### Let's check Stationary of the Time Series using Augmented Dickey-Fuller Test

#### Null hypothesis is that datasets are not stationary

```python
print('Augmented Dickey-Fuller Test on Albeny Rainfall Data')
dftest = adfuller(albeny_data['Rainfull'],autolag='AIC')
dftest
```

```python
print('Augmented Dickey-Fuller Test on Mississippi Rainfull Data')
dftest = adfuller(mississippi_data['Rainfull'],autolag='AIC')
dftest
```

#### Since p-value<0.05, we reject the null hypothesis
#### Hence both the datasets are stationary.

## 2. Deep Learning (LSTM)

### AJMER

```python
#Scaled the data

scaler = MinMaxScaler()
scaler.fit(train_a)
scaled_train = scaler.transform(train_a)
scaled_test = scaler.transform(test_a)
```

```python
# define generator
n_input = 24
n_features = 1
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=1)
```

```python
# define model
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()model.summary()
```

```python
# fit model
model.fit_generator(generator,epochs=50)
```

```python
model.history.history.keys()
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```

```python
test_predictions = []

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))

for i in range(len(test_a)):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model.predict(current_batch)[0]
    
    # store prediction
    test_predictions.append(current_pred) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
test_predictions
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
```

```python
test = test_a
test['Predictions'] = true_predictions
```

```python
test
```

```python
test.plot(figsize=(12,8))
```

#### Evaluation

```python
from statsmodels.tools.eval_measures import rmse

error = rmse(test['Rainfull'], test['Predictions'])
print(error)
```
